```python
#getting genres
def get_genres(i):
    l=[]
    for j in eval(i):
        l.append(j["name"])
    return l


#getting keywords
def get_keywords(i):
    l=[]
    for j in eval(i):
        l.append(j["name"])
    return l

#getting top 5 cast name
def get_cast(i):
    l=[]
    s=0
    for j in eval(i):
        s+=1
        l.append(j["name"])
        if s==5:
            break
    return l
```

# using WordNet to to make similar words a single word

```python
from nltk.stem import WordNetLemmatizer
```

```python
"""
#need to download all this packages
nltk.download("punkt")
nltk.download("wordnet")
nltk.download("omw-1.4")
nltk.download('punkt_tab')
"""
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
movies_tag_set["tags"].iloc[0].split()
```

```python
l = []
for i in movies_tag_set["tags"]:
    words = word_tokenize(i)
    lemmatized_words =" ".join([lemmatizer.lemmatize(word) for word in words])
    l.append(lemmatized_words)
len(l)
```

```python
movies_tag_set["tags"] = l
```

# vectorizing tags

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000,stop_words = "english")
```

```python
vectorized_tags = cv.fit_transform(movies_tag_set["tags"]).toarray()
```

# getting similarity between each and every tags

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectorized_tags)
```

```python
similarity.shape
```

# exectuion of main function

```python
def top_five(movie):
    idx=movies_tag_set[movies_tag_set["original_title"]==movie].index[0]
    listidx=list(enumerate(similarity[idx]))
    similar = sorted(listidx, key = lambda x:x[1], reverse = True)
    
    temp = []
    for i in similar[0:6]:
        MT=movies_tag_set.iloc[i[0]]["original_title"]
        MID=movies_tag_set.iloc[i[0]]["id"]
        temp.append((MID,MT))
    return temp
        
```

```python
def recommendations(movie):
    try:
        return top_five(movie)
    except IndexError :
        return "No such Movie Exist...!!"
```

```python
recommendations("Avatar")
```

```python
import random

# Function to generate random reviews
def generate_reviews(num_reviews=100):
    movie_reviews = dict()
    movie_id = []
    reviews = []
    for id in movies["id"]:
        
        for _ in range(num_reviews):
            sentiment = random.choice(['positive', 'negative'])
        
            if sentiment == 'positive':
                review = random.choice(positive_templates)
            else:
                review = random.choice(negative_templates)
        
            reviews.append(review)
            movie_id.append(id)
        
    movie_reviews = {"movie_id":movie_id,
                     "reviews":reviews}
    return movie_reviews

# Generate 50,000 reviews
reviews = generate_reviews()

# Create DataFrame
import pandas as pd
df = pd.DataFrame(reviews)

# Display the first few reviews

```
